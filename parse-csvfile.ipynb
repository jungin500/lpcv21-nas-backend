```python
import pandas as pd
from IPython.display import display

filename = 'csvout/Graph1_21-07-19  20-40.csv'
data = pd.read_csv(filename, skiprows=[0]).rename(columns={
    'Vbus(V)': 'Vbus',
    'Ibus(A)': 'Ibus',
    'Power(W)': 'Power',
    'Vd+(V)': 'Vdp',
    'Vd-(V)': 'Vdm',
    'energy(Wh)': 'Wh',
    'temperature(��)': 'temp'
}).interpolate()
```

```python
display(data)
```

```python
Watts = data['Vbus'] * data['Power']
display(Watts)
```

```python
print("Items: ", len(Watts))

print("Mean:", Watts.mean())
print("Variance:", Watts.var())
print("StdDev:", Watts.std())
```
